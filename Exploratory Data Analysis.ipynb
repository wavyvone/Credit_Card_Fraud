# Exploratory Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib import rcParams

# figure size in inches
rcParams['figure.figsize'] = 11, 10
```

```python
data = pd.read_csv("./Data/card_transdata.csv")
data.head()
```

```python
data["fraud"].value_counts().plot(kind='bar')
plt.title("Fraud Value Counts")
plt.xticks([0.0, 1.0], ["False", "True"])
```

```python
sns.pairplot(data.sample(2500))
```

```python
sns.heatmap(data.corr(), annot=True, fmt='.4f', cmap='Blues')
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.histplot(data=np.log(data.sample(2500)[data["fraud"] < 0.5]), x="distance_from_home", color="blue", label="Not Fraud", kde=True)
sns.histplot(data=np.log(data.sample(2500)[data["fraud"] > 0.5]), x="distance_from_home", color="red", label="Fraud", kde=True)
plt.xlabel("Distance From Home (Log)")
plt.legend() 
plt.show()
```

```python
sns.histplot(data=np.log(data.sample(2500)[data["fraud"] < 0.5]), x="distance_from_last_transaction", color="blue", label="Not Fraud", kde=True)
sns.histplot(data=np.log(data.sample(2500)[data["fraud"] > 0.5]), x="distance_from_last_transaction", color="red", label="Fraud", kde=True)
plt.xlabel("Distance From Last Transaction (Log)")
plt.legend() 
plt.show()
```

```python
sns.histplot(data=np.log(data.sample(2500)[data["fraud"] < 0.5]), x="ratio_to_median_purchase_price", color="blue", label="Not Fraud", kde=True)
sns.histplot(data=np.log(data.sample(2500)[data["fraud"] > 0.5]), x="ratio_to_median_purchase_price", color="red", label="Fraud", kde=True)
plt.xlabel("Ratio To Median Purchase Price (Log)")
plt.legend() 
plt.show()
```

```python
count_data = {
    "True - Fraud": [],
    "False - Fraud": [],
    "True - No Fraud": [],
    "False - No Fraud": []
}

data_point = "repeat_retailer"
data_point_label = "Repeat Retailer"

for column in ['repeat_retailer', 'used_chip', 'used_pin_number', 'online_order']:
    fraud_counts = data[data["fraud"] > 0.5][column].value_counts()
    count_data["False - Fraud"].append(fraud_counts[0])
    count_data["True - Fraud"].append(fraud_counts[1])
    
    not_fraud_counts = data[data["fraud"] < 0.5][column].value_counts()
    count_data["False - No Fraud"].append(not_fraud_counts[0])
    count_data["True - No Fraud"].append(not_fraud_counts[1])
    
```
